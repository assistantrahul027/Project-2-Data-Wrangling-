```python
sns.boxplot(x= 'cnt', data=combined_dataset)     #checking outliers of Combined_dataset
plt.title('Boxplot of Values')                   # Add title and labels

plt.xlabel('Values')                                         
plt.show()
```

```python
Q1 = np.percentile(combined_dataset['cnt'], 25, method='midpoint')              # Calculating percentiles and IQR
Q3 = np.percentile(combined_dataset['cnt'], 75, method='midpoint')
IQR = Q3 - Q1
print(IQR)
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR                   # Define outlier boundaries


# Identify outliers

outliers = combined_dataset[(combined_dataset['cnt'] < lower_bound) | (combined_dataset['cnt'] > upper_bound)]
print(outliers)
```

# Correlation Matrix



```python
# Calculate correlation matrix
correlation_matrix = combined_dataset.corr(numeric_only=True)

print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))             # Plot heatmap of the correlation matrix using matplotlib lib.

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)

plt.title('Correlation Heatmap')
plt.show()                                                   
```
